1. $x>-1$  and  $x<1$  
2. $x>-6$  and  $x<6$  
3. $x>0$   and  $x<5$  


Niezbędne pakiety do wykonania zadania:

```python
def fun(x):
    return x**2 + 5
def rysuj(x,przedzial):
    plt.plot(x,fun(x),"r",label = 'funkcja y = x**2 + 5')
    plt.xlabel("Argumenty x")
    plt.ylabel("Zbiór wartości y")
    plt.grid()
    plt.legend()
    plt.title("Wykres funkcji w przedziale {}".format(przedzial))
    plt.show()
```
